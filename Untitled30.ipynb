```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime

# ==================== 設定（已替換為您指定的6支美股） ====================
TICKERS = ["AAPL", "MSFT", "JNJ", "PG", "JPM", "XOM"]
START_DATE = "2020-01-01"
END_DATE = datetime.today().strftime('%Y-%m-%d')

# 目標權重（根據各股票特性設計，涵蓋成長、穩定、防御、周期型，總和=1）
TARGET_WEIGHTS = {
    "AAPL": 0.20,   # 科技高成長
    "MSFT": 0.20,   # 科技穩定增長
    "JNJ":  0.15,   # 醫療保健防御型
    "PG":   0.15,   # 消費必需品防御型
    "JPM":  0.15,   # 金融周期型
    "XOM":  0.15    # 能源周期型
}

# ==================== 修復：下載資料（關閉自動復權，保留Adj Close） ====================
print("正在下載資料...")
# 關鍵修復：添加 auto_adjust=False，消除警告+保留復權價列
data = yf.download(TICKERS, start=START_DATE, end=END_DATE, auto_adjust=False)['Adj Close']
returns = data.pct_change().dropna()

print(f"資料期間：{data.index[0].date()} 至 {data.index[-1].date()}")
print(f"資料筆數：{len(data)}")

# ==================== 假設目前持倉（總值10萬美元，刻意製造偏差以展示再平衡） ====================
current_holdings = pd.Series({
    "AAPL": 25000,   # 目前超配成長股
    "MSFT": 18000,   # 略低
    "JNJ":  12000,   # 明顯低配防御股
    "PG":   16000,
    "JPM":  17000,
    "XOM":  12000
})

current_weights = current_holdings / current_holdings.sum()

# ==================== 顯示結果 ====================
print("\n=== 目前權重 vs 目標權重（AAPL/MSFT/JNJ/PG/JPM/XOM） ===")
comparison = pd.DataFrame({
    "目前權重": current_weights,
    "目標權重": pd.Series(TARGET_WEIGHTS),
    "偏差": current_weights - pd.Series(TARGET_WEIGHTS)
})
print(comparison.round(4))

# 簡單建議
print("\n=== 初步再平衡建議（決策輸出） ===")
deviation = comparison["偏差"]
print("建議：")
for ticker, dev in deviation.items():
    if dev > 0.02:
        print(f"   賣出 {ticker}（權重過高 {dev:.1%}）")
    elif dev < -0.02:
        print(f"   買入 {ticker}（權重過低 {abs(dev):.1%}）")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
from scipy.optimize import minimize

# ==================== 核心函数：风险平价权重计算 ====================
def calculate_risk_parity_weights(returns):
    """计算风险平价权重：各资产风险贡献相等"""
    cov = returns.cov() * 252  # 年化协方差矩阵
    n_assets = len(cov)
    init_weights = np.array([1/n_assets]*n_assets)  # 初始等权
    bounds = tuple((0, 1) for _ in range(n_assets)) # 权重0-1

    def risk_budget_objective(weights):
        vol = np.sqrt(np.dot(weights.T, np.dot(cov, weights)))
        risk_contrib = (weights * np.dot(cov, weights)) / vol
        return np.sum((risk_contrib - risk_contrib.mean())**2)

    cons = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    result = minimize(risk_budget_objective, init_weights, bounds=bounds, constraints=cons)
    return pd.Series(result.x, index=returns.columns)

# ==================== 核心函数：再平衡 + 交易成本计算 ====================
def rebalance_portfolio(current_weights, target_weights, portfolio_value, cost_rate=0.001):
    """
    执行再平衡，计算交易金额与成本
    cost_rate: 美股平均交易成本 0.1%（符合專案要求）
    """
    weight_diff = target_weights - current_weights
    trade_value = abs(weight_diff * portfolio_value)
    total_cost = np.sum(trade_value) * cost_rate
    new_value = portfolio_value - total_cost
    return {
        "weight_diff": weight_diff,
        "trade_value": trade_value,
        "estimated_cost": total_cost,
        "new_portfolio_value": new_value
    }

# ==================== 設定（已替換為AAPL/MSFT/JNJ/PG/JPM/XOM） ====================
TICKERS = ["AAPL", "MSFT", "JNJ", "PG", "JPM", "XOM"]
START_DATE = "2020-01-01"
END_DATE = datetime.today().strftime('%Y-%m-%d')
REBALANCE_FREQUENCY = "3M"  # 每季再平衡

# ==================== 下载数据 ====================
print("正在下载数据...")
data = yf.download(TICKERS, start=START_DATE, end=END_DATE, auto_adjust=False)['Adj Close']
returns = data.pct_change().dropna()
print(f"数据期间：{returns.index[0].date()} 至 {returns.index[-1].date()}")

# ==================== 回测逻辑 ====================
print("\n开始 Backtesting...")
portfolio_value = 100000  # 初始本金10万美元
current_weights = pd.Series([1/len(TICKERS)]*len(TICKERS), index=TICKERS)  # 初始等权
results = []
rebalance_dates = returns.resample(REBALANCE_FREQUENCY).last().index

for date in rebalance_dates:
    # 截取历史数据计算目标权重（风险平价）
    hist_returns = returns.loc[:date]
    target_weights = calculate_risk_parity_weights(hist_returns)

    # 执行再平衡
    rebalance_result = rebalance_portfolio(current_weights, target_weights, portfolio_value)

    # 更新持仓
    current_weights = target_weights
    portfolio_value = rebalance_result["new_portfolio_value"]

    # 计算当期收益
    period_daily_returns = returns.loc[date:].iloc[:63]  # 季度约63个交易日
    if len(period_daily_returns) > 0:
        strat_return = period_daily_returns.dot(current_weights).mean()
        portfolio_value *= (1 + strat_return)

    # 记录结果
    results.append({
        "Date": date,
        "Portfolio_Value": round(portfolio_value, 2),
        "Rebalance_Cost": round(rebalance_result["estimated_cost"], 2)
    })

# 转换为数据表
backtest_df = pd.DataFrame(results).set_index("Date")

# ==================== 绩效评估指标 ====================
daily_ret = backtest_df["Portfolio_Value"].pct_change().dropna()
final_value = backtest_df["Portfolio_Value"].iloc[-1]
total_return = (final_value / 100000 - 1) * 100
sharpe = (daily_ret.mean() / daily_ret.std()) * np.sqrt(252) if len(daily_ret) > 0 else 0
max_dd = ((backtest_df["Portfolio_Value"] / backtest_df["Portfolio_Value"].cummax()) - 1).min() * 100
total_cost = backtest_df["Rebalance_Cost"].sum()

# ==================== 输出结果 ====================
print(f"\n=== 回测结果（每季风险平价再平衡）===")
print(f"初始本金：100,000 美元")
print(f"最终资产价值：{final_value:,.0f} 美元")
print(f"总报酬率：{total_return:.2f}%")
print(f"Sharpe 比率：{sharpe:.3f}")
print(f"最大回撤：{max_dd:.2f}%")
print(f"总交易成本：{total_cost:,.0f} 美元")
print(f"单次平均交易成本：{(total_cost/len(rebalance_dates)):.2f} 美元")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
import plotly.graph_objects as go

# ==================== 核心函数（内置） ====================
def rebalance_portfolio(current_weights, target_weights, portfolio_value, cost_rate=0.001):
    weight_diff = target_weights - current_weights
    trade_value = weight_diff * portfolio_value
    total_trade_amount = np.sum(np.abs(trade_value))
    total_cost = total_trade_amount * cost_rate
    cost_ratio = total_cost / portfolio_value
    new_value = portfolio_value - total_cost

    suggested_trades = []
    for ticker, diff in trade_value.items():
        if abs(diff) < 100:
            continue
        if diff < 0:
            suggested_trades.append(f"【賣出】{ticker} : {abs(round(diff,2))} USD")
        else:
            suggested_trades.append(f"【買入】{ticker} : {round(diff,2)} USD")

    return {
        "weight_diff": weight_diff,
        "trade_value": trade_value,
        "estimated_cost": total_cost,
        "cost_ratio": cost_ratio,
        "new_portfolio_value": new_value,
        "suggested_trades": suggested_trades
    }

# ==================== 工具參數設定（已替換為AAPL/MSFT/JNJ/PG/JPM/XOM） ====================
TICKERS = ["AAPL", "MSFT", "JNJ", "PG", "JPM", "XOM"]
total_value = 100000  # 總資產（美元）
transaction_cost = 0.001  # 0.1%交易成本

# 目標權重（同上）
target_weights = pd.Series({
    "AAPL": 0.20,
    "MSFT": 0.20,
    "JNJ":  0.15,
    "PG":   0.15,
    "JPM":  0.15,
    "XOM":  0.15
})

# 模擬目前持倉（刻意製造偏差）
current_holdings = pd.Series({
    "AAPL": 25000,
    "MSFT": 18000,
    "JNJ":  12000,
    "PG":   16000,
    "JPM":  17000,
    "XOM":  12000
})

current_weights = current_holdings / current_holdings.sum()

# ==================== 执行再平衡 ====================
result = rebalance_portfolio(current_weights, target_weights, total_value, transaction_cost)

# ==================== 输出結果（完全符合專案「可操作決策輸出」要求） ====================
print("✅ 再平衡計算完成！（投資組合再平衡助理）")
print(f"預估交易成本：{result['estimated_cost']:,.0f} USD ({result['cost_ratio']:.2%})")
print(f"再平衡後資產：{result['new_portfolio_value']:,.0f} USD")
print(f"建議交易筆數：{len(result['suggested_trades'])}")
print("\n📋 建議買賣清單（可直接執行）：")
for trade in result['suggested_trades']:
    print(trade)

# 互動式圖表（Streamlit / Jupyter 皆可顯示）
fig = go.Figure()
fig.add_bar(x=current_weights.index, y=current_weights, name="目前權重", marker_color="#FF6B6B")
fig.add_bar(x=target_weights.index, y=target_weights, name="目標權重", marker_color="#4ECDC4")
fig.update_layout(
    barmode='group',
    title="目前權重 vs 目標權重（AAPL/MSFT/JNJ/PG/JPM/XOM）",
    xaxis_title="股票標的",
    yaxis_title="權重",
    height=500
)
fig.show()
```
